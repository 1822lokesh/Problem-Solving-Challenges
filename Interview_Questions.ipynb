1. Write a Python function to find all unique pairs of numbers in the array that add up to the target sum.


    * Example Input:

        purchases = [1, 2, 3, 4, 3, 5, 1], target = 6

    * Expected Output:

        [(1, 5), (2, 4), (3, 3)]

```python
def find_unique(arr, target):

  seen = set()
  unique_pairs = set()

  for num in arr:
    item = target - num

    if item in seen:
      pair = tuple(sorted((num, item)))
      unique_pairs.add(pair)
    seen.add(num)

  return list(unique_pairs)

# test the fuction
purchases = [1, 2, 3, 4, 3, 5, 1]
target = 6
result = find_unique(purchases, target)
print(f"Final result : {result}")
```
