**Question 1: The Palindrome Checker**

**Problem Statement:** Write a function to determine if a given string is a palindrome. A palindrome is a word that reads the same backward as forward (e.g., "madam" or "racecar").

**Requirements:**

* Input: A string.

* Output: Boolean (True or False).

* Constraint: The check must be case-insensitive ("Madam" should be True).

* Bonus: Can you do it in a single line of code?

```python
def palindrome_check(s):
  chars = list(s)
  left = 0
  right = len(s) - 1
  while left < right:
    chars[left], chars[right] = chars[right], chars[left]
    left += 1
    right -= 1
  return "".join(chars)
```
