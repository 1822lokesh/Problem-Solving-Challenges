**Question 1: Find the Factorial**

**Problem Statement:** Write a function to calculate the factorial of a given non-negative integer n. 
A factorial ($n!$) is the product of all positive integers less than or equal to $n$.

Requirements:
* Input: An integer (e.g., 5).
* Output: The factorial result (e.g., 120).
* Constraint: Use a while loop or a for loop (do not use the math.factorial library).
* Handle the edge case where $0! = 1$.

```python
def factorial_num(num):
    if num < 0:
          return "Factorial is not defined for negative numbers."
    fact = 1
    for n in range(1, num+1):
            fact = fact * n
    return fact

factorial_num(5)
```

**Question 3: Find the Second Largest Number**

**Problem Statement:**

Given a list of numbers, find the second largest value in that list.

**Requirements:**

* Input: [10, 20, 4, 45, 99]

* Output: 45

* Constraint: Do not use list.sort().

* Handle the case where the list might contain duplicate numbers.

```python
nums = [10, 20, 4, 45, 90]
first = 0
second = 0
for n in nums:
    if n > first:
        first = n
    elif n > second and n < first:
        second = n
print(second)

    
```
