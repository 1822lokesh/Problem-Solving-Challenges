**Question 1: Find the Factorial**

**Problem Statement:** Write a function to calculate the factorial of a given non-negative integer n. 
A factorial ($n!$) is the product of all positive integers less than or equal to $n$.

Requirements:
* Input: An integer (e.g., 5).
* Output: The factorial result (e.g., 120).
* Constraint: Use a while loop or a for loop (do not use the math.factorial library).
* Handle the edge case where $0! = 1$.

```python
def factorial_num(num):
    if num < 0:
          return "Factorial is not defined for negative numbers."
    fact = 1
    for n in range(1, num+1):
            fact = fact * n
    return fact

factorial_num(5)
```

**Question 3: Find the Second Largest Number**

**Problem Statement:**

Given a list of numbers, find the second largest value in that list.

**Requirements:**

* Input: [10, 20, 4, 45, 99]

* Output: 45

* Constraint: Do not use list.sort().

* Handle the case where the list might contain duplicate numbers.

```python
def second_largest(nums):
    if len(nums) < 2:
        return "No second largest element"

    largest = None
    second_largest = None

    for num in nums:
        if largest is None or num > largest:
            if largest != num:
                second_largest = largest
            largest = num

        elif num != largest and (second_largest is None or num > second_largest):
            second_largest = num

    if second_largest is None:
        return "No second largest element"

    return second_largest



numbers = [10, 20, 30, 4, -6, 95, 80, 75]
print(second_largest(numbers))  
```

**Question 4: Count Word Occurrences**

**Problem Statement:**

Write a function that takes a sentence (string) and returns a dictionary where keys are the words and values are how many times they appeared.

**Requirements:**

* Input: "the cat sat on the mat"

* Output: {'the': 2, 'cat': 1, 'sat': 1, 'on': 1, 'mat': 1}

* Use the .split() method to break the string into words.

* Ensure it is case-insensitive ("The" and "the" should be the same).

```python
def word_count(text):
    words = text.split()
    freq = {}
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq
print(word_count("the cat sat on the mat"))
```

**Question 5: List Intersection**

**Problem Statement:**

Given two lists, return a new list that contains only the elements that are common to both lists (the intersection).

**Requirements:**


* Input: list1 = [1, 2, 3, 4], list2 = [3, 4, 5, 6]

* Output: [3, 4]

* Use a list comprehension to solve this in one line.

* Ensure there are no duplicates in the final list.

```python
list1= [1,2,3,4]
list2= [3,4,5,6]
common = []
for i in list1:
    for j in list2:
        if i == j:
            common.append(i)
print(common)
```

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

intersection = list({x for x in list1 if x in list2})

print(intersection)  
```
