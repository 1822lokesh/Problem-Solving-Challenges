**Question 1: The Leap Year Checker**

**Problem Statement:**

Write a function that determines if a given year is a leap year. A year is a leap year if it is divisible by 4, but not by 100, unless it is also divisible by 400.

**Requirements:**

* Input: An integer (e.g., 2024).

* Output: Boolean (True or False).

* Constraint: Use nested if statements or a single line with and/or operators.

* Handle the logic carefully for "century years" like 1900 (False) and 2000 (True).

```python
def leap_year(year):
  if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    return True
  else:
    return False

year = int(input("enter the years"))
print(leap_year(year))
```

**Question 2: List Slicing & Rotation**

**Problem Statement:**
Given a list, "rotate" it to the right by one position. This means the last element becomes the first, and all other elements shift right.

**Requirements:**

* Input: [1, 2, 3, 4, 5]

* Output: [5, 1, 2, 3, 4]

* Constraint: Use list slicing ([:]) to achieve this in a single line.

* The function should work regardless of the list length.

```python
list1 = [1, 2, 3, 4, 5]
print(list1[0 : -1] )
```
