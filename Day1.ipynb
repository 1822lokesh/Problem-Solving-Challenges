**Question 1: The Vowel Counter**

**Problem Statement:** Write a function that counts the number of vowels in a given string. The function should be case-insensitive.


**Requirements:**

* Input: A string (e.g., "Hello World").

* Output: An integer count of vowels (a, e, i, o, u).

* Constraint: Use a for loop and a simple if condition.

* Handle uppercase and lowercase letters correctly (e.g., 'A' and 'a' both count).

```python
def vowles_count(string):
  count = 0
  vowels = ['a','e','i','o','u','A','E','I','O','U']
  for i in string:
    for j in vowels:
      if i == j:
        count+= 1
  print(f"The number of vowels presented in this sentence is {count} letters")

user = input("Enter your sentence:")
vowles_count(user)
```

**Question 2: List Deduplication**

**Problem Statement:**  You are given a list of integers that contains duplicates. Your task is to remove the duplicates while keeping the original order of the elements.

**Requirements:**

* Input: [1, 2, 2, 3, 4, 1, 5]

* Output: [1, 2, 3, 4, 5]

* Constraint: Do NOT use set() (because set loses order).

* Use a new list and check for existence before appending.

```python
def remove_dup(lst):
  new_list = []
  for i in lst:
    if i not in new_list:
      new_list.append(i)
  return new_list

original_list = [1,2,4,4,8,2,5,5,3,3,3,6,7,7,9,9,9,8,11,10,10]
remove_dup(original_list)
```

**Question 3: String Reversal (Without Slicing)**

**Problem Statement:** Python allows string reversal easily with text[::-1]. For this interview question, you must write a function to reverse a string without using slicing.

**Requirements:**

* Input: "Python"

* Output: "nohtyP"

* Use a loop or the reversed() built-in function combined with join.

* Return the new string.

**1. Slicing Method**

* **Syntax:** s[::-1]

* **How it works:** This uses Python's "step" slicing. The -1 tells Python to start from the end and move backward to the start, one character at a time.

* **Why use it:**  It is the fastest method because it happens at the C-level (the language Python is built in).
* **Time Complexity:**  $O(n)$

* **Space Complexity:**$O(n)$ (It creates one new reversed string).

**2. Built-in join() + reversed()**
* **Syntax:** "".join(reversed(s))
* **How it works:** reversed(s) creates an "iterator" (a pointer) that walks backward. "".join() then stitches those characters together into a new string.
* **Why use it:** It is very readable and memory-efficient. It doesn't copy the string until the very last moment.

* **Time Complexity:**  $O(n)$
* **Space Complexity:** $O(n)$

**3. The Accumulator Loop**
* **Syntax:** for char in s: rev_str = char + rev_str

* **How it works:** It takes each character from the original and "glues" it to the front of a new string.

* **Why use it:** Itâ€™s great for beginners to understand how iteration works, but it is slow for large strings because Python has to rebuild the string object every time you add a letter.
* **Time Complexity:** $O(n^2)$ (Due to repeated string copying).
* **Space Complexity:** $O(n)$

**4. Two-Pointer Swap**
* **Syntax:** while left < right: (with list(s) and temp variable).

* **How it works:** You convert the string to a list, place a "pointer" at both ends, and swap the characters as they move toward the middle.

* **Why use it:** This is the "Interview Method." It shows you understand memory and algorithms. Since Python strings are immutable, you must convert to a list first.

* **Time Complexity:** $O(n)$

* **Space Complexity:** $O(n)$ (One list and one final string).

```python
def reverse_string(s):
  rev_str=""
  for char in s:
    rev_str = char + rev_str
  return rev_str

input_str = input("Enter your string to reverse:")
reverse_string(input_str)
```

```python
def reverse_two_pointers(s):
    # Convert to list (because strings are immutable)
    chars = list(s)
    left = 0
    right = len(chars) - 1

    while left < right:
        chars[left], chars[right] = chars[right], chars[left]
        left += 1
        right -= 1
    return "".join(chars)

print(reverse_two_pointers("Software"))
```

**Question 4: Find the Maximum (The Logic Check)**

**Problem Statement:** Write a function that finds the largest number in a list of integers without using the built-in max() function and without sorting the list.

**Requirements:**

* Input: [10, 50, 20, 5]

* Output: 50

* Initialize a variable (e.g., current_max) and iterate through the list to compare.

* Handle negative numbers correctly (don't just initialize current_max to 0).

```python
def largest_num(numbers):
  if not numbers:
    return None
  max_num = numbers[0]
  for num in numbers:
    if num > max_num:
      max_num = num
  return max_num

print(largest_num([20,10,5,60,30,40,70,15]))
print(largest_num([-10, -5, -20, -3, -1]))
```

**Question 5: Dictionary Square Map**

**Problem Statement:** Create a function that takes an integer n and generates a dictionary where the keys are numbers from 1 to n and the values are the square of the keys.


**Requirements:**

* Input: n = 5

* Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

* Use a for loop or a dictionary comprehension.

* Return the final dictionary.

```python
def square_map(n):
    result = {}
    for i in range(1, n + 1):
        result[i] = i * i
    return result

print(square_map(10))
```

```python
def square_maps(n):
  return {i : i * i for i in range(0,n+1)}
square_maps(10)
```
