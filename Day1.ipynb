Question 1: The Vowel Counter
Problem Statement: Write a function that counts the number of vowels in a given string. The function should be case-insensitive.


**Requirements:**

* Input: A string (e.g., "Hello World").

* Output: An integer count of vowels (a, e, i, o, u).

* Constraint: Use a for loop and a simple if condition.

* Handle uppercase and lowercase letters correctly (e.g., 'A' and 'a' both count).

```python
def vowles_count(string):
  count = 0
  vowels = ['a','e','i','o','u','A','E','I','O','U']
  for i in string:
    for j in vowels:
      if i == j:
        count+= 1
  print(f"The number of vowels presented in this sentence is {count} letters")

user = input("Enter your sentence:")
vowles_count(user)
```
